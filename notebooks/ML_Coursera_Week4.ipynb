```python
import numpy as np

theta0 = -10
theta1 = 20
theta2 = 20

def compute_z(x1, x2):
  return theta0 + theta1*x1 + theta2*x2

def sigmoid(z):
    return 1 / (1 + np.exp(-z))

print(sigmoid(compute_z(0, 0)))
print(sigmoid(compute_z(1, 0)))
print(sigmoid(compute_z(0, 1)))
print(sigmoid(compute_z(1, 1)))
```
